```python
def hist_all(img):
    
    color = ('b', 'g', 'r')
    for i, colour in enumerate(color):
        histr = cv2.calcHist([img], [i], None, [256], [0, 256])
        plt.plot(histr, color=colour)
        plt.xlim([0, 256])
    
    img = ''
    plt.title('Histogram')
```

```python
hist_all(blue_bricks)
```

```python
hist_all(dark_horse)
```

```python
def hist_horse(img):
    
    color = ('b', 'g', 'r')
    for i, colour in enumerate(color):
        histr = cv2.calcHist([img], [i], None, [256], [0, 256])
        plt.plot(histr, color=colour)
        plt.xlim([0, 50])
        plt.ylim([0, 500000])
        
    plt.title('Histogram for blue bricks')
```

```python
hist_horse(dark_horse)
```

```python
gorilla = cv2.imread('DATA/gorilla.jpg', 0)
```

```python
def display(img, cmap=None):
    fig = plt.figure(figsize=(12, 10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap)
```

```python
display(gorilla, cmap='gray')
```

```python
gorilla.shape
```

```python
hist_values = cv2.calcHist([gorilla], channels=[0], mask=None, histSize=[256], ranges=[0, 256])
```

```python
plt.plot(hist_values)
```

```python
eq_gorilla= cv2.equalizeHist(gorilla)
display(eq_gorilla, cmap='gray')
```

```python
hist_values = cv2.calcHist([eq_gorilla], channels=[0], mask=None, histSize=[256], ranges=[0, 256])
```

```python
plt.plot(hist_values)
```
