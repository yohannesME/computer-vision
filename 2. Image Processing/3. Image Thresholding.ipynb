```python
img = cv2.imread('DATA/crossword.jpg', 0)
```

```python
plt.imshow(img, cmap = 'gray')
```

```python
def show_pic(img):
    fig = plt.figure(figsize=(20, 20))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
show_pic(img)
```

```python
ret, threshold_binary = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
```

```python
show_pic(threshold_binary)
```

```python
ret, threshold_binary = cv2.threshold(img, 200, 255, cv2.THRESH_BINARY)
```

```python
show_pic(threshold_binary)
```

```python
ret, threshold_binary = cv2.threshold(img, 150, 255, cv2.THRESH_BINARY)
```

```python
show_pic(threshold_binary)
```

```python
threshold_adaptive = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 8)
show_pic(threshold_adaptive)
```

```python
blended = cv2.addWeighted(src1=threshold_binary, alpha=0.6, src2=threshold_adaptive, beta=0.4, gamma=0)
```

```python
show_pic(blended)
```

```python
threshold_gaussian = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 8)
show_pic(threshold_gaussian)
```
